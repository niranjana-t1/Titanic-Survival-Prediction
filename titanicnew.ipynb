# Titanic Survival Prediction
## Problem Definition
This project aims to predict whether a passenger survived the Titanic disaster using machine learning techniques.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
titanic_data=pd.read_csv('train.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.shape
```

```python
titanic_data.info()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data = titanic_data.drop(columns='Cabin',axis=1)
```

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

```python
print(titanic_data['Embarked'].mode())
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python
titanic_data['Embarked'] = titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0])
```

```python
titanic_data.isnull().sum()
```

```python
#Data Analysis
```

```python
titanic_data.describe()
```

```python
titanic_data['Sex'] = titanic_data['Sex'].str.strip().str.lower()
titanic_data['Embarked'] = titanic_data['Embarked'].str.strip().str.upper()
```

```python
#encoding
```

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
titanic_data['Sex'] = titanic_data['Sex'].map({'male':0, 'female':1})
titanic_data['Embarked'] = titanic_data['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
titanic_data.head()
```

```python
X=titanic_data.drop(columns=["PassengerId","Name","Ticket","Survived"],axis=1)
Y=titanic_data["Survived"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy score
```

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of training data: ", training_data_accuracy)
```

```python
#accuracy on test data
```

```python
X_test_prediction= model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy score of test data: ",test_data_accuracy)
```

```python
#Confusion Matrix
```

```python
Y_pred = model.predict(X_test)

cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
#Random Forest
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)

print("RF Accuracy:", rf.score(X_test, Y_test))
```

```python
#result
```

```python

columns = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']

# Your new passenger data (list of lists)
new_passenger = [[1, 1, 29, 0, 1, 92.0, 2]]

# Convert to DataFrame
new_passenger_df = pd.DataFrame(new_passenger, columns=columns)

# Predict
prediction = model.predict(new_passenger_df)

# Convert 0/1 to Yes/No
result = "Yes" if prediction[0] == 1 else "No"
print("Will this passenger survive?", result)
```
